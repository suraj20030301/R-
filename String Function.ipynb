```r
#string slicing
str1<- as.character (readline("Enter name :"))
n<-as.integer(readline("Starting point :"))
m<-as.integer(readline("ending point :"))

print(substr(str1, n, m))
```

```r
##Grep funtion check patter given as input
str<-"Hello! Every one this is R programming trial"
lst<-list("C++","Java","Python","C","Excel")
print(grep("C",lst))
print(grep("t",str))
```

Above you can see it return the index of given input where it is exist, but in str you can see output is 1 because the string is in 1 index of str

```r
#Replace the string using user input
str<-"Hello! Every one this is R programming trial"
lst<-list("C++","Java","Python","C","Excel")
print(sub("R","Python",str))
print(sub("C","B",lst))
```

```r
#Concatenation of strings, if separate is given then it will follow
str1<-"I am pursuing BCA from Srinath Universiy"
str2<-"My name is Suraj Kumar Gupta"
print(paste(str2,str1))
print(paste("Hello!",str2,str1,sep = ","))
```

```r
#String split using split function, it will convert string in list type
str2<-"My name is Suraj Kumar Gupta"
spl<-strsplit(str2," ")
typeof(spl)
class(spl)
print(spl)
```

```r
#Upper and Lower Case
nam<-"Suraj Kumar Gupta"
print(toupper(nam))
print(tolower(nam))
```
