```r
#Math Functions
a<- -25.36

#(i) Absolute value
print(abs(a))
```

```r
#(ii) Squre root, if user input is +ve result will show on output, if input is -ve is will return NaN
a<- -81
b<- 81
print(sqrt(a))
print(sqrt(b))
```

```r
#ceiling function, it will return smaller value of grater than the input
x<- 7.2
print(ceiling (x))
```

```r
#floor function, it is similer to ceiling but it return gretest value of smaller than the input
x<- 7.2
print(floor(x))
```

```r
#it will truncate the input same as floor
x<- 88.6
y<- 88.4
print(trunc(x))
print(trunc(y))
```

```r
#it has two parameters one is value and second is no of place for round figure
a<-12.353566
print(round(a,digits=3))
```

```r
#Trigonometric values
p<-90
print(sin(p))
print(cos(p))
print(tan(p))
```

```r
#Logarithm function, it is a natural log function base value is  e.
print(log(20))

#it is use as given variable base
print(log10(100))
print(log2(8))
```

```r
#it will return exponent of given input

print(exp(100))
print(exp(5))
```
