```r
#Data Frame and its Operstions
#Creating Data Frame
df<-data.frame(roll_no=c(401,402,403,404),Sname=c("Suresh","Suraj","Shruti","Rahul"),attendance=c(90,70,95,85))
df
```

```r
#Define Data as vector and making Data Frame
uid<-c(4453,0908,2231,3434,4523)
name<-c("Samresh","Rohit","Vivek","Rajesh","Amit")

df<-data.frame(uid,name)
df
```

```r
#Accessing Data Frame
#print specific column using index
df[1]

#print diffirent columns using index
df[c(1,2)]

#print column using columns name
df['uid']
```

```r
#print column name using $sign

df$uid

#print value from a column using $sign

df$uid[2]
```

```r
#Operations of Data Frame
summary(df)

#can summarize single column also
summary(df$uid)
```

```r
#Adding row or columns in existing data frame
#Adding column
product<-c("Soap","Surf","Nuts","biscuits","Chocolates")
df<-cbind(df,product)
df

#Adding Row
data<-list(8753,"Deepak","Haldiram")
df<-rbind(df,data)
df
```

```r
#Removing Data from Data Frame
#Removing Row
df<-df[-c(3),]
df
#Removing Column
df<-df[,-c(3)]
df

#Removing Both row and column
df<-df[-c(5),-c(1)]
```

```r
#finding number of rows and columns
df<-data.frame(name<-c("Ram","Siya","Laxman"),age<-c(134,124,130))
df
nrow(df)  #for row
ncol(df)  #for column

```
