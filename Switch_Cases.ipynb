```r
#using Arithmetic oprestions

print("This is Suraj's Calculator of two numbers only\n")
print("---------------------------\n")
num1<-as.integer(readline("1st number is: "))
operation<-readline("Operation sign: ")
num2<-as.integer(readline("2nd number is: "))


res3<-switch (operation,
  "+" = num1+num2,
  "-" = num1-num2,
  "*" = num1*num2,
  "/" = num1/num2,
  "invalid sign"
)
print(res3)
```
