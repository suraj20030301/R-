```r
#different ways to declare vectors
```

```r
#1
a<-1:10
print("Type 1: ")
print(a)

#2
b<-seq(1,10)
print("Type 2: ")
print(b)
c<-seq(1,10,by = 3)
print("Type 2 with gaps: ")
print(c)
```

```r
#Generally it is declare using C function
#vector is homogenius data stucture, all elements should be same data type like integer,character,logical etc.
#by default it is in numeric data type

vect1<-c(1,2,3,4,5,6,67)
vect1
typeof(vect1)
class(vect1)
print("")

#Charcter
vect2<-c("Suraj","Kumar","Gupta")
vect2
typeof(vect2)
class(vect2)
print("")

#logical
a<-10
b<-7
vect3<-c(a<b,a>b,a<=b,a>=a)
vect3
typeof(vect3)
class(vect3)
```

```r
#if you'll try to forcely add different data types elements it convert all to, its first element

vect4<-c("Suraj","Kumar","Gupta",12,TRUE,45.555)
vect4
typeof(vect4)
class(vect4)
```

```r
#Operarions of vectors

#Arithmetic operations
x<-c(10:20)
y<-c(seq(100,200,by = 10))
x;y

print("")
print(x+y)
print(y-x)
print(y*x)
print(y/x)
print(y%%x)
```

```r
#Merge operations

p<-c(x,y)
print(p)
```

```r
#indexing
x[1] #it will display first value of vector
x[-1] #it will display all values expext 1 index(mentioned index)

x[-5]
```

```r
#Duplicate indexing & range index both are same,   you can create duplicate index also but first you have to create index vector
x
a<-c(3,5,10,2,1)
print(x[a])
```

```r
#Out of order index    if user demand out of range index then it will print NA

x
a<-c(3,5,10,2,1,20,21)
print(x[a])
```
