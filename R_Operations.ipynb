```r
.num1<-as.integer(readline(prompt="First Number: "))
.num2<-as.integer(readline(prompt="Second Number: "))

#Arithmetic Operations
print(.num1+.num2)
print(.num1-.num2)
print(.num1*.num2)
print(.num1/.num2)
print(.num1%%.num2)
print(.num1%/%.num2)
print(.num1^.num2)
```

```r
#Relational Operations
print(.num1<.num2)
print(.num1>.num2)
print(.num1<=.num2)
print(.num1>=.num2)
print(.num1!=.num2)
print(.num1==.num2)
```

```r
#Logical Operations
var=TRUE
var.=FALSE

print(var&&var.)
print(var||var.)
print(var!=var.)

vec1<-c(TRUE,FALSE,TRUE)
vec2<-c(TRUE,TRUE,FALSE)

print(vec1&vec2)
print(vec1|vec2)
```

```r
#Special Operations
x=2:10
print(x)

print(5 %in% x)
print(2 %*% x)
```
