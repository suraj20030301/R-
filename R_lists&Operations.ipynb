```r
#List and its operations.
#List is similar to Vector but it can contain diffrent datatypes.

list1<-list("Suraj",23,50+2i,40L)
list1
typeof(list1)
class(list1)
```

```r
vec<-c(1,2,15)
vec_char<-c("Suraj","Kumar","Gupta")
vec_log<-c(TRUE,FALSE,FALSE,TRUE)
typeof(vec)
typeof(vec_char)
typeof(vec_log)
class(vec)
class(vec_char)
class(vec_log)
print("")
list2<-list(vec,vec_char,vec_log)
typeof(list2)
class(list2)
```

```r
#print using index

list2
list2[1]
list2[2]
list2[3]
```

```r
#list in list

list3<-list(c("Suraj","Sahil","Ajay"),matrix(c(20,21,23),nrow = 1),list("DS","C++","C#"))
list3
print("")

#Named the columns
names(list3)<-c("StudentName","Age","Subjects")
list3

#Can Search with name
print("")
list3["Age"]
```

```r
#Unlist a list and it will act as vector
x<-unlist(list3)
x
```

```r
#Merge List
a<-list(1,2,32)
b<-list(4,71,90)
mergeList<-list(a,b)
mergeList
```

```r
#manipulate list
list3[1]<-"Std"
list3
```

```r
#Can drop a data from list
list3[1]<-NULL
list3
```
